```python
# Lets select the Numerical Columns from the Data for check cluster .
x=crop.iloc[:,0:7].values
#Create Data Frame
x_data=pd.DataFrame(x,columns=np.arange(0,7))
print(x_data.head())
print(x_data.shape)
```

```python
# lets determine the Optimum Number of Clusters within the Dataset
from sklearn.cluster import KMeans
plt.rcParams['figure.figsize'] = (10, 4)

wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters = i,)
    km.fit(x_data)
    wcss.append(km.inertia_)

# lets plot the results
plt.plot(range(1, 11), wcss,color='red')
plt.title('The Elbow Method', fontsize = 20)
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
km = KMeans(n_clusters = 4)
no_clu=km.fit_predict(x_data)
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x[no_clu==0,0],x[no_clu==0,1],color='red')
plt.scatter(x[no_clu==1,0],x[no_clu==1,1],color='yellow')
plt.scatter(x[no_clu==2,0],x[no_clu==2,1],color='green')
plt.scatter(x[no_clu==3,0],x[no_clu==3,1],color='blue')
```

```python
# lets find out the which crop gone which cluster
a =crop['label']
no_clu = pd.DataFrame(no_clu)
z = pd.concat([no_clu, a], axis = 1)
z = z.rename(columns = {0: 'cluster'})

# lets check the Clusters of each Crops
print("Lets check the Results After Applying the K Means Clustering Analysis \n")
print("Crops in First Cluster:", z[z['cluster'] == 0]['label'].unique())
print("---------------------------------------------------------------")
print("Crops in Second Cluster:", z[z['cluster'] == 1]['label'].unique())
print("---------------------------------------------------------------")
print("Crops in Third Cluster:", z[z['cluster'] == 2]['label'].unique())
print("---------------------------------------------------------------")
print("Crops in Forth Cluster:", z[z['cluster'] == 3]['label'].unique())
```

## Create Model

```python
x=crop.iloc[:,0:7]
y=crop.iloc[:,-1]
print("Shape of X = ", x.shape)
print("Shape of Y = ", y.shape)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)

print("Shape of X_train = ",x_train.shape)
print("Shape of x_test = ",x_test.shape)
print("Shape of y_train = ",y_train.shape)
print("Shape of Y_test = ",y_test.shape)
```

```python
# Data Scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
# Data Transform
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
# Salect data Type
from sklearn.compose import make_column_selector
num_col=make_column_selector(dtype_include='number')
cat_col=make_column_selector(dtype_include='number')
```

```python
# columns_transformer pipeline
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
trf=make_column_transformer(
  (make_pipeline(scaler),num_col))
```

```python
# lets create a Predictive Model
from sklearn.linear_model import LogisticRegression
pipe=make_pipeline(trf,LogisticRegression())
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
```

```python
# Check Confusion matrix first
from sklearn.metrics import classification_report, confusion_matrix
plt.rcParams['figure.figsize'] = (10, 10)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True,)
plt.title('Confusion Matrix for Logistic Regression', fontsize = 15)
plt.show()
```

```python
# lets print the Classification Report also
cr = classification_report(y_test, y_pred)
print(cr)
```
